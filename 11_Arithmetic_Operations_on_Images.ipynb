#### Image Addition
<p>You can add two images by OpenCV function, cv2.add() or simply by numpy operation, res = img1 + img2. Both images should be of same depth and type, or second image can just be a scalar value.
<p>**Note:** There is a difference between OpenCV addition and Numpy addition. OpenCV addition is a saturated operation while Numpy addition is a modulo operation.
<p>For example, consider below sample:


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

x = np.uint8([250])
y = np.uint8([10])

print cv2.add(x,y) # 250+10 = 260 => 255
print x+y # 250+10 = 260 % 256 = 4
```

It will be more visible when you add two images. OpenCV function will provide a better result. So always better stick
to OpenCV functions.
